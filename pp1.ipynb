```python
df_main=pd.merge(area_harvested[['Area', 'Item', 'Year', 'Value']].rename(columns={'Value': 'Area_harvested'}),
                yield_data[['Area', 'Item', 'Year', 'Value']].rename(columns={'Value': 'Yield'}),
                on=['Area', 'Item', 'Year'],
                how='inner'
            )
```

>Feature preparation

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import  r2_score, accuracy_score, mean_absolute_error, mean_squared_error
```

```python
enc_area=LabelEncoder()
enc_item=LabelEncoder()
```

```python
feature_names = ['Area_encoded', 'Item_encoded', 'Year', 'Area_harvested', 'Yield']
```

```python
df_main['Area_encoded'] = enc_area.fit_transform(df_main['Area'])
df_main['Item_encoded'] = enc_item.fit_transform(df_main['Item'])
```

```python
path1="/Users/ameen/Documents/PROJECTS_NEW/Crop/Data/cleaned_data.csv"
df_main.to_csv(path1,index=False)
```

```python

X = pd.DataFrame(df_main[feature_names])
y = df_main['Production']
```

>Train moodel

```python
model = RandomForestRegressor(
                n_estimators=200,  
                max_depth=20,      
                min_samples_split=5,
                min_samples_leaf=2,
                n_jobs=-1  
            )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

>Metrics

```python

r2_score = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mean_squared_error(y_test, pred))
CV_scores = cross_val_score(model, X, y, cv=5)
            
```

```python
print("r2_score",r2_score)
print("Mean squared error",mse)
print("root mean squared error",rmse)
print('cross validation score',CV_scores)
```
